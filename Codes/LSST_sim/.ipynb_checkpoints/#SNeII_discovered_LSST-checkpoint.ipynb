```python
c_light=299792.458   # in km/s
c_AA=299792458*1.0e10# in AA/s
alpha_IMF=1.35       # Salpeter IMF: Salpeter et al. 1955 ApJ 121: 161
z_max=1.0            # redshift maximum simulatio
z_sim=np.arange(0.01,z_max,0.01)
fow_LSST=9.62        # field of view LSST in degrees^2
delta_omega=fow_LSST*(math.pi/180)**2      # field of view LSST in str
h_erg = 6.63e-27     # Planck constant (erg.s)
h = 0.70             # dimensionless Hubble constant
H0 = 100.0 * h       # Hubble constant in Km/s/Mpc
d_h=c_light/H0   # Hubble distance in Mpc   (70 km/s / Mpc)

#LSST limiting magnitude
mag_lim={'u':23.60,'g':24.83,'r':24.38,'i':23.92,'z':23.35,'y':22.44} #mag lim 1 expo grizY
#mag_lim_10=[26.1,27.4,27.5,26.8,26.1,24.9]   #mag lim stacked images after 10 years

```

```python
#cosmology
omega_m=0.30
omega_k=0.0
omega_lambda=0.70
def e_z_LCDM(z):
    return (1.0/math.sqrt(omega_m*((1+z)**3)+ omega_k*((1+z)**2) + omega_lambda))
    e_z_int_LCDM, e_z_int_err_LCDM = integrate.quad(e_z_LCDM,0.,z)
e_z_distri=np.vectorize(e_z_LCDM)


def r_com_LCDM(z):                   # We define comoving distance for flat universe in Mpc
    return (d_h*integrate.quad(e_z_LCDM,0.,z)[0])
r_com_distri=np.vectorize(r_com_LCDM)

def d_lum_LCDM(z):                   # We define luminosity distance for flat universe in Mpc
    return (r_com_LCDM(z)*(1+z))
d_lum_distri=np.vectorize(d_lum_LCDM)
def mu_LCDM(z):                   # We define modulus distance for flat universe
    return (5.0 * np.log10(d_lum_LCDM(z)*10**6/10))
d_mod_distri=np.vectorize(mu_LCDM)

def madau_time(t):                   # We define t9 as Madeau et al 1998
    return (13*pow((1+t),-1.5))

#Initial Mass Function
#REFERENCE: Salpeter et al. 1955 ApJ 121: 161
def salpeter(M):
    return (pow(M,-(1+alpha_IMF)))
def Mass_tot(M):
    return (M*pow(M,-(1+alpha_IMF)))

#Let's define different Star Formation Rate
#The number of SN is define as the number of stars that explodes as SNeII (KII) * SFR
#REFERENCES : Madau et al. 1998 MNRAS, 297, L17
#            Horiuchi et al. 2011 ApJ, 738, 154H
#            Cole et al. 2001  MNRAS, 326, 255C
a=3.4
b=-0.3
c=-3.5
ro=0.016*0.73
eta=-10
B=2**(1-a/b)
C=2**((b-a)/c)*5**(1-b/c)
def SFR_Horiuchi(z):
    return (ro*((1+z)**(a*eta)+((1+z)*1.0/B)**(b*eta)+((1+z)*1.0/C)**(c*eta))**(-0.1))
def SNR_Horiuchi(z):
    return (KII*SFR_Horiuchi(z))

a1=0.0166 
b1=0.1848 
c1=1.9474 
d1=2.6316 
def SFR_Cole(z):
    return (h*(a1+b1*z)/(1+(z/c1)**d1))
def SNR_Cole(z):
    return (KII*SFR_Cole(z))   

a2=0.015 
b2=2.7 
c2=2.9
d2=5.6 
def SFR_Madau(z):
    return (a2*((1+z)**b2/(1+((1+z)/c2)**d2)))
def SNR_Madau(z):
    return (KII*SFR_Madau(z)) 

def dust(z):
    if z<3.3:
        f_dust=0.95-0.28*z
    else:
        f_dust=0.02
    return f_dust 

def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
```

# 1. Preliminaries

## 1.1 LSST filter responses



```python
x_func={}
s_x={}
dem_x={}
lambda_filter={}
filterlist = ('u', 'g', 'r', 'i', 'z', 'y')
lsst_effl={'u':3654.9,'g':4800.3,'r':6222.0,'i':7540.6,'z':8682.1,'y':9916.6}
filtercolors = {'u':'b', 'g':'c', 'r':'g', 'i':'orange', 'z':'r', 'y':'m'}

fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')
for f in filterlist:
    #DL the filter bandpass
    trans_file=np.loadtxt('Filters/LSST_'+f+'.dat')
    lambda_filt=trans_file[:,0]    #in AA
    s_x[f]=trans_file[:,1]
    #interpolate filter bandpass
    x_func[f]=interpolate.interp1d(lambda_filt,s_x[f])
    dem_x[f]=integrate.simps(x_func[f](lambda_filt)/lambda_filt,lambda_filt) 
    
    ax1.plot(lambda_filt, x_func[f](lambda_filt), color=filtercolors[f], lw=2, label=r'\textbf{LSST %s}' % (f))
    lambda_filter[f]=trans_file[:,0]
ax1.set_xlabel(r'\textbf{Wavelength [\AA]}')
ax1.set_ylabel(r'\textbf{Throughput [0-1]}')
ax1.set_title(r'\textbf{LSST Throughput Curves}')
ax1.legend(loc=(0.8, 0.6), fancybox=True, fontsize='smaller')
ax1.grid(True)
#plt.savefig(os.path.join('LSSTfilters.png'), format='png')
plt.show()
```

## 1.2 K-corrections

We use [Dessart et al. 2013](http://esoads.eso.org/abs/2013MNRAS.433.1745D) models to measure K-corrections between bands. The K-corrections are necessary as the Universe is in expansion, one photon emitted in one filter in the rest-frame will be observed from Earth in another filter. For example a low redshift one photon emitted in r band will be observed in r band, however, at z>0.1 and z$<$0.3, the photon emitted in r will be received in i band.

```python
class Kcorrection(object):
    def __init__(self):
        self.array = np.empty((len(filterlist),len(filterlist),len(z_sim)))
        self.interpolate ={}
    
    def __getitem__(self,f1f2z):
        f1,f2,z=f1f2z
        return self.array[filterlist.index(f1),filterlist.index(f2),z]

    def __setitem__(self,f1f2z,value):
        f1,f2,z=f1f2z
        self.array[filterlist.index(f1),filterlist.index(f2),z]=value
    
    def interpol(self,f1,f2):
        if f1+f2 not in self.interpolate:
            self.interpolate[f1+f2]=interpolate.interp1d(z_sim,self[f1,f2,:])  
        return self.interpolate[f1+f2]
```

```python
#take Dessart' model 50 days after explosion. We calculate only the K-correction at 50 days after the explosion
ascii=pd.read_table('theoretical_model/m15_du_sch_mlt3_FeC_mix0p4_18.fl',delim_whitespace=True,header=None)
lambda_model=ascii[0].values
flux_model=ascii[1].values
F_spec_model_0=interpolate.interp1d(lambda_model,flux_model)

Kcor=Kcorrection()
m_lsst_za={}
m_lsst_0={}
for i,zs in enumerate(z_sim):
    lambda_model_obs=lambda_model*(1+zs)
    flux_model_obs=flux_model/(1+zs)
    F_spec_model_z=interpolate.interp1d(lambda_model_obs,flux_model_obs)
    
    for f in filterlist:
        #comparison of F_spec_model_z and F_spec_model_0  
        #Int(Fobs*S*l)
        #-2.5*log10( Int(Fobs*S*l)   ) + 2.5*log10( Int(S/l) ) 
        #Comparison of F_spec_model_z(SDSS) and F_spec_model_0(CSP)
        #Int(Frest*S*l)
        #-2.5*log10( Int(Frest*S*l)) + 2.5*log10( Int(S/l) )

        n_lsst_za=integrate.simps(F_spec_model_z(lambda_filter[f])*x_func[f](lambda_filter[f])*lambda_filter[f],lambda_filter[f])   
        m_lsst_za[f]=-2.5*(np.log10(n_lsst_za/dem_x[f]))-48.60+2.5*(np.log10(c_AA))
        n_lsst_0=integrate.simps(F_spec_model_0(lambda_filter[f])*x_func[f](lambda_filter[f])*lambda_filter[f],lambda_filter[f])   
        m_lsst_0[f]=-2.5*(np.log10(n_lsst_0/dem_x[f]))-48.60+2.5*(np.log10(c_AA))

    for j,f1 in enumerate(filterlist):
        for k,f2 in enumerate(filterlist):
            Kcor[f1,f2,i]=m_lsst_za[f1]-m_lsst_0[f2]
```

## 1.3 Here we construct a K-correction term from all bands to r-band

```python
Kxr=np.zeros(len(z_sim))
for i,zs in enumerate(z_sim):
    #select the closest filter to r at a certain redshift
    lambda_eff=lsst_effl['r']*(1+zs)    
          
    ind_X = np.where(np.array([x for x in lsst_effl.values()]) ==find_nearest([x for x in lsst_effl.values()],lambda_eff) )[0][0]    
    Kxr[i]=Kcor[filterlist[ind_X],'r',i]

Kxr_func=interpolate.interp1d(z_sim,Kxr)
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')

for j,f1 in enumerate(filterlist):
    ax1.plot(z_sim,Kcor.interpol(f1,'r')(z_sim),filtercolors[f1])
ax1.plot(z_sim,Kxr_func(z_sim),'black')
ax1.set_xlabel(r'\textbf{Redshift}')
ax1.set_ylabel(r'\textbf{Kxr}')
plt.ylim(-2,2)
ax1.legend((r'\textbf{u to r',r'\textbf{g to r}',r'\textbf{r to r}',r'\textbf{i to r}',r'\textbf{z to r}',r'\textbf{y to r}',r'\textbf{Final KC}'),loc=3,markerscale=0.5,prop={'size':15},ncol=2) 
plt.show()

#Final Kcorrections fron x to r in Kx_r and Kxr_func
# r->r if zs <= 0.105:
# i->r if zs > 0.105 and zs <= 0.303:
# z->r if zs > 0.303 and zs <= 0.494:
# y->r if zs > 0.494:
```

# 2. Calculating SNII rate

We will basically apply the approach by [Forster et al. 2006](http://adsabs.harvard.edu/abs/2006MNRAS.368.1893F)/[Strogler et al. 2004](http://adsabs.harvard.edu/abs/2004ApJ...613..200S) adapted to SNe II following [Botticella et al. 2012](http://adsabs.harvard.edu/abs/2012A%26A...537A.132B)


The rate of SNe II per unit time per unit comoving volume (RII) is given by the star formation rate (SFR) per unit time per unit comoving volume convolved with the number of stars crearted that will explode as SNe II (which we will estimate assuming a [Salpeter et al. (1955)](http://adsabs.harvard.edu/abs/1955ApJ...121..161S) IMF and their progenitors are stars from 8 to 25 solar masses).

### R$_{II}$=K$_{II}$ x SFR

## 2.1 Calculating the fraction of stars that explode as SN II

### K$_{II}$= integrale(IMF) over the II (mass between 8-25) / integrale (IMF) over all the masses

```python
mass_min=8
mass_max=25

KII=integrate.quad(salpeter,mass_min,mass_max)[0]/(integrate.quad(Mass_tot,0.1,100.))[0]
print(KII)
```

## 2.2 Calculating SFR

As an example, we will use 3 different SFR evolutions with redshift: [Cole et al. (2001)](http://adsabs.harvard.edu/abs/2001MNRAS.326..255C), [Horiuchi et al. (2011)](http://adsabs.harvard.edu/abs/2011ApJ...738..154H), and [Madau et al. (2014)](http://adsabs.harvard.edu/abs/2014ARA%26A..52..415M)

```python
SFR_Horiuchi_vect=vectorize(SFR_Horiuchi)
SFR_Cole_vect=vectorize(SFR_Cole)
SFR_Madau_vect=vectorize(SFR_Madau)
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')

ax1.plot(z_sim,np.log10(SFR_Horiuchi_vect(z_sim)),'g',z_sim,np.log10(SFR_Cole_vect(z_sim)),'k',z_sim,np.log10(SFR_Madau_vect(z_sim)),'r')
ax1.legend((r'\textbf{Horiuchi}',r'\textbf{Cole}',r'\textbf{Madau}'))
ax1.set_xlabel(r'\textbf{redshift}')
ax1.set_ylabel(r'\textbf{log(SFR) [h$^{-1}$ M$_{solar}$ yr$^{-1}$ Mpc$^{-3}$]}')
plt.show()
```

```python
#Now let's plot the SN rate

SNR_Horiuchi_vect=vectorize(SNR_Horiuchi)
SNR_Cole_vect=vectorize(SNR_Cole)
SNR_Madau_vect=vectorize(SNR_Madau)
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')

ax1.plot(z_sim,np.log10(SNR_Horiuchi_vect(z_sim)),'g',z_sim,np.log10(SNR_Cole_vect(z_sim)),'k',z_sim,np.log10(SNR_Madau_vect(z_sim)),'r')
ax1.legend((r'\textbf{Horiuchi}',r'\textbf{Cole}',r'\textbf{Madau}'))
ax1.set_xlabel(r'\textbf{redshift}')
ax1.set_ylabel(r'\textbf{log SN}')
ax1.set_title(r'\textbf{SN rate per unit time per unit comoving volume}')

plt.show()
```

# 3. Example with  a random SN II: SN II 2005J

## 3.1 At which redshift SN 2005J would not be detected with LSST?
We will first take SN II 2005J as a reference and assume that all SNe II are like 05J. We simulate its r-band light curve at different redshifts and see for which z it would be below the LSST detections limit


```python
#Data 2005J
JD_explosion=2453382.78
JD_ref=2400000
z_cmb=4183./c_light
err_z_cmb=150./c_light

SN_photo=pd.read_table('Data/2005J_r.dat',delim_whitespace=True,header=0)
SN_epoch=SN_photo['MJD']+JD_ref-JD_explosion
mag_r=SN_photo['Mag']
dmag_r=SN_photo['DMag']

#Absolute mag 2005J
M_SN_r=mag_r-5*np.log10(d_lum_distri(z_cmb))-25
dM_SN_r=np.sqrt(dmag_r**2+(np.array(err_z_cmb)*(5*(1+np.array(z_cmb))*1.0/(np.array(z_cmb)*(1+np.array(z_cmb)*1.0/2)*np.log10(10))))**2)

#Figure absolute mag
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')

ax1.plot(SN_epoch,M_SN_r,'rs',ms=8, linestyle='--') 

ax1.set_ylim(ax1.get_ylim()[::-1])
ax1.set_title(r'\textbf{SN 2005J}')
ax1.set_ylabel(r'\textbf{Absolute magnitude [mag]}')
ax1.set_xlabel(r'\textbf{Epoch since explosion [days]}')
plt.show()


#Calcul of the apparent magnitude at different redshift
m_sn=np.zeros(shape=(len(SN_epoch),len(z_sim)))
for t,snep in enumerate(SN_epoch):
    for z,zs in enumerate(z_sim):
        m_sn[t,z]=M_SN_r[t]+d_mod_distri(zs)+Kxr_func(zs)

fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')        
for z,zs in enumerate(z_sim):
    ax1.plot(SN_epoch,m_sn[:,z],label=r'\textbf{z=%s}'%zs) 
    if min(m_sn[:,z]) > mag_lim['r']:
        print('Not visible from z='+str(zs))
        break
ax1.legend(loc=(1.0, 0.25), fancybox=True, fontsize='smaller',markerscale=0.5,prop={'size':15},ncol=4) 
ax1.set_ylim(ax1.get_ylim()[::-1])
ax1.axhline(mag_lim['r'], color='k', linestyle='--')
ax1.set_ylabel(r'\textbf{Apparent magnitude [mag]}')
ax1.set_xlabel(r'\textbf{Epoch since explosion [days]}')
ax1.set_title(r'\textbf{SN 2005J at several redshifts}')
plt.show()
```

## 3.2 How many SNe II (2005J-like) would the LSST discover in 1 year?

First, we compare our apparennt magnitude $m(t,z)$ with limiting magnitude $m_{lim}$ of the LSST camera to obtain $\Delta_{t}(z)$,  the probability of detecting (the peak of) a SN at a given redshift 

```python
t_min=[0]*len(z_sim)
t_max=[0]*len(z_sim)
t_max_2=[0]*len(z_sim)
delta_t_redshift=[0]*len(z_sim) #delta t corrected by redshift
epoch_model_z=[0]*len(SN_epoch)

for i,zs in enumerate(z_sim):
    epoch_model_z=SN_epoch*(1+zs)  #1+z factor for time dilation
    if min(m_sn[:,i][:])<mag_lim['r']:
        t_min[i]=epoch_model_z[min(np.where(m_sn[:,i]<=mag_lim['r'])[0])]
        o=set(np.where(epoch_model_z>=t_min[i])[0])   #index of the last t>t_min
        p=set(np.where(m_sn[:,i]<=mag_lim['r'])[0])        #index at which we see the SN
        q=(list(o-p))                                 #all the indeces where we see the peak
        if q==[]:
            t_max[i]=epoch_model_z[max(np.where(m_sn[:,i]<=mag_lim['r'])[0])]
            delta_t_redshift[i]=t_max[i]-t_min[i]
        else:
            t_max[i]=epoch_model_z[min(q)]
            t_max_2[i]=epoch_model_z[max(q)]
            delta_t_redshift[i]=(t_max[i]-t_min[i])+(max(epoch_model_z)-t_max_2[i])#le SBO+ rest of curve
```

Now we estimate the number of detected SNe II per unit of redshift, $dN/dz$. From Equations 2 and 5 of Forster et al. (2006)

```python
dV_dz_dW=np.zeros(len(z_sim))

#############
survey_length=1.0   #in years
#############

#The number of detected SNe II per unit redshift interval
n_Horiuchi=[0]*len(z_sim)   
n_Cole=[0]*len(z_sim)      
n_Madau=[0]*len(z_sim)      

for i,zs in enumerate(z_sim):
    dV_dz_dW[i]=(pow(r_com_distri(zs),2)*d_h*e_z_distri(zs))#dV(z) is the volume per unit redshift being surveyed

    #total observing time multiplied by the probability of detecting a SN
    n_Horiuchi[i]=SNR_Horiuchi(zs)/(1.+zs)*dV_dz_dW[i]*delta_omega*((delta_t_redshift[i])/365.25)*survey_length #In yr (SNR in yr)     #dn/dz is the SN per unit of redshift, comovil volume, solid angle and fied of view
    n_Cole[i]=SNR_Cole(zs)/(1+zs)*dV_dz_dW[i]*delta_omega*((delta_t_redshift[i])/365.25)*survey_length #In yr (SNR in yr)          #dn/dz is the SN per unit of redshift, comovil volume, solid anfle and fied of view
    n_Madau[i]=SNR_Madau(zs)/(1+zs)*dV_dz_dW[i]*delta_omega*((delta_t_redshift[i])/365.25)*survey_length #In yr (SNR in yr)        #dn/dz is the SN per unit of redshift, comovil volume, solid anfle and fied of view
                               
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')
    
ax1.plot(z_sim,n_Horiuchi,'g',z_sim,n_Cole,'k',z_sim,n_Madau,'r')
ax1.set_title(r'\textbf{\# SNe II per unit of dz: SN2005J-like }')
ax1.set_xlabel(r'\textbf{redshift}')
ax1.set_ylabel(r'\textbf{\# SN }')
plt.legend((r'\textbf{Horiuchi}',r'\textbf{Cole}',r'\textbf{Madau}'))
plt.show()

# We integrate for all the redshift to have the number of SN 
print('Num. SNe Horiuchi: '+str(int(np.sum(n_Horiuchi))))
for i,zs in enumerate(z_sim):
    if n_Horiuchi[i] == 0.0:
        print('No SNe II up to z='+str(zs))
        break
        
print('Num. SNe Cole:     '+str(int(np.sum(n_Cole))))
for i,zs in enumerate(z_sim):
    if n_Cole[i] == 0.0:
        print('No SNe II up to z='+str(zs))
        break
        
print('Num. SNe Madau:     '+str(int(np.sum(n_Madau))))
for i,zs in enumerate(z_sim):
    if n_Madau[i] == 0.0:
        print('No SNe II up to z='+str(zs))
        break
```

# 4. Example with a model of a 'typical' SN II

Instead of assuming that all SNe II are like SN 2005J, we will use the spectral model of [Dessart et al. (2013](http://adsabs.harvard.edu/abs/2013MNRAS.433.1745D) which is anchored to SN II 1999em. 

This SN is a bit fainter compared to SN 2005J ($M_{99em}=-16.6$; $M_{05J}=-17.2$), so we expect to find a distribution with a cutout at lower redshifts.

```python
##This loads all models in lam,flux_model
model=pd.read_table('theoretical_model/m15_du_sch_mlt3_FeC_mix0p4_list',delim_whitespace=True,header=None)
filename=model[0].values
epoch=model[1].values

lam_model=len(epoch)*[0]
flux_model=len(epoch)*[0]

SN99em=pd.read_table('Data/1999em_R.dat',delim_whitespace=True,header=0)
ep_99em=SN99em['MJD']-51476.5
mag_99em=SN99em['Mag']
F_99em=interpolate.interp1d(ep_99em,mag_99em)

mapp_99em=len(filename)*[0]
Mabs_99em=len(filename)*[0]
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')

z_99em=0.002392
#We will plot the  model from Dessart et al. at different epochs. However first we need to normalize the spectrum 
#using the SN photometry (mangling).
for i,fname in enumerate(filename):
    if i ==0:
        ascii=pd.read_table('theoretical_model/'+fname+'.fl',delim_whitespace=True,header=None)
        lam_model[i]=ascii[0].values
        flux_model[i]=ascii[1].values
        #mangling
        F_spec_model_z=interpolate.interp1d(lam_model[i]/(1+z_99em),flux_model[i]*(1+z_99em))
        n_obs=integrate.simps(F_spec_model_z(lambda_filter['r'])*x_func['r'](lambda_filter['r'])*lambda_filter['r'],lambda_filter['r'])
        m_obs_r=-2.5*np.log10(n_obs/dem_x['r'])-48.60+2.5*(np.log10(c_AA))
        factor=10**(-0.4*(F_99em(epoch[i])-m_obs_r))
                
    ascii=pd.read_table('theoretical_model/'+fname+'.fl',delim_whitespace=True,header=None)
    lam_model[i]=ascii[0].values
    flux_model[i]=ascii[1].values*factor
    
    F_spec_model_z=interpolate.interp1d(lam_model[i],flux_model[i])
    num=integrate.simps(F_spec_model_z(lambda_filter['r'])*x_func['r'](lambda_filter['r'])*lambda_filter['r'],lambda_filter['r'])
    mapp_99em[i]=-2.5*np.log10((num/dem_x['r']))-48.60+2.5*(np.log10(c_AA))
    Mabs_99em[i]=mapp_99em[i]-mu_LCDM(z_99em)

    ax1.plot(lam_model[i],flux_model[i])

ax1.set_ylabel(r'\textbf{Flux [ergs cm$^{-2}$ s$^{-1}$ \AA$^{-1}$]}')    
ax1.set_xlabel(r'\textbf{Rest frame wavelength [\AA]}')    

ax1.set_xlim(1000,15000)
plt.show()
```

```python
#Apparent magnitude of the model at different epochs and redshifts
m_sn_model=np.zeros(shape=(len(epoch),len(z_sim)))
for i,snep in enumerate(epoch):
    for t,zs in enumerate(z_sim):
        m_sn_model[i,t]=Mabs_99em[i]+d_mod_distri(zs)+Kxr_func(zs)

fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')
for i,zs in enumerate(z_sim):
    ax1.plot(epoch,m_sn_model[:,i],label=r'\textbf{z=%s}'%zs)
    if min(m_sn_model[:,i]) > mag_lim['r']:
        print('Not visible from z='+str(zs))
        break

ax1.set_ylim(ax1.get_ylim()[::-1])
ax1.axhline(mag_lim['r'], color='k', linestyle='--')
ax1.set_xlabel(r'\textbf{Epochs since explosion [days]}')
ax1.set_ylabel(r'\textbf{Apparent magnitude [mag]}')
ax1.set_title(r'\textbf{Luc at several redshifts')
ax1.legend(loc=(1.0, 0.25), fancybox=True, fontsize='smaller',markerscale=0.5,prop={'size':15},ncol=4) 
plt.show()
```

## 4.2 How many SNe II (Dessart-like) would the LSST discover in 1 year?

First, we compare our $m(t,z)$ with $m_{lim}$ of the LSST camera to obtain $\Delta_{t}(z)$,  the probability of detecting (the peak of) a SN at a given redshift 

```python
t_min=[0]*len(z_sim)
t_max=[0]*len(z_sim)
t_max_2=[0]*len(z_sim)
delta_t_redshift=[0]*len(z_sim) #delta t corrected by redshift
epoch_model_z=[0]*len(epoch)

for i,zs in enumerate(z_sim):
    epoch_model_z=epoch*(1+zs)  #1+z factor for time dilation
    if min(m_sn_model[:,i][:])<mag_lim['r']:
        t_min[i]=epoch_model_z[min(np.where(m_sn_model[:,i]<=mag_lim['r'])[0])]
        o=set(np.where(epoch_model_z>=t_min[i])[0])              #index of the last t>t_min
        p=set(np.where(m_sn_model[:,i]<=mag_lim['r'])[0])        #index at which we see the SN
        q=(list(o-p))                                            #all the indeces where we see the peak
        if q==[]:
            t_max[i]=epoch_model_z[max(np.where(m_sn_model[:,i]<=mag_lim['r'])[0])]
            delta_t_redshift[i]=t_max[i]-t_min[i]
        else:
            t_max[i]=epoch_model_z[min(q)]
            t_max_2[i]=epoch_model_z[max(q)]
            delta_t_redshift[i]=(t_max[i]-t_min[i])+(max(epoch_model_z)-t_max_2[i])
```

```python
dV_dz_dW=np.zeros(len(z_sim))

#############
survey_length=1.0   #in years
#############

#The number of detected SNe II per unit redshift interval
n_Horiuchi_mod=[0]*len(z_sim)   
n_Cole_mod=[0]*len(z_sim)      
n_Madau_mod=[0]*len(z_sim)      
for i,zs in enumerate(z_sim):
    dV_dz_dW[i]=(pow(r_com_distri(zs),2)*d_h*e_z_distri(zs))                 #dV(z) is the volume per unit redshift being surveyed

    n_Horiuchi_mod[i]=SNR_Horiuchi(zs)/(1.+zs)*dV_dz_dW[i]*delta_omega*((delta_t_redshift[i])/365.25)*survey_length #In yr (SNR in yr)     #dn/dz is the SN per unit of redshift, comovil volume, solid angle and fied of view
    n_Cole_mod[i]=SNR_Cole(zs)/(1+zs)*dV_dz_dW[i]*delta_omega*((delta_t_redshift[i])/365.25)*survey_length #In yr (SNR in yr)          #dn/dz is the SN per unit of redshift, comovil volume, solid anfle and fied of view
    n_Madau_mod[i]=SNR_Madau(zs)/(1+zs)*dV_dz_dW[i]*delta_omega*((delta_t_redshift[i])/365.25)*survey_length #In yr (SNR in yr)        #dn/dz is the SN per unit of redshift, comovil volume, solid anfle and fied of view                               
    
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')    
ax1.plot(z_sim,n_Horiuchi_mod,'g',z_sim,n_Cole_mod,'k',z_sim,n_Madau_mod,'r')
ax1.set_title(r'\textbf{\# SNe II per unit of dz: Dessart et al. }')
ax1.set_xlabel(r'\textbf{redshift}')
ax1.set_ylabel(r'\textbf{\# SN }')
plt.legend((r'\textbf{Horiuchi}',r'\textbf{Cole}',r'\textbf{Madau}'))
plt.show()

# We integrate for all the redshift to have the number of SN 
print('Num. SNe Horiuchi: '+str(int(np.sum(n_Horiuchi_mod))))
for i,zs in enumerate(z_sim):
    if n_Horiuchi_mod[i] == 0.0:
        print('No SNe II from z='+str(zs))
        break
        
print('Num. SNe Cole:     '+str(int(np.sum(n_Cole_mod))))
for i,zs in enumerate(z_sim):
    if n_Cole_mod[i] == 0.0:
        print('No SNe II from z='+str(zs))
        break
        
print('Num. SNe Madau:     '+str(int(np.sum(n_Madau_mod))))
for i,zs in enumerate(z_sim):
    if n_Madau_mod[i] == 0.0:
        print('No SNe II from z='+str(zs))
        break
print('As expected the distributions cut at lower redshift than for SN 2005J as the models are fainter.')
```

# 5. Example with a CSP-like simulated sample of SNe II

Finally, we will use a a sample of 10,000 SNe, simulated using MCMC on real parameters of CSP SNe II. We will put each at 100 random redshifts form 0.01 to 1.00, so we will have 1,000,000 SNe.

Note that what we also change here is that we will consider the magnitude at the end of the plateau phase. This is becuase in SNII cosmology, we do not standardize the magnitude at peak but either the magnitude at around the center of the plateau, i.e. at ~50 days (for spectroscopic methods), or the length/brightness-decline of the plateu (for photometric methods).

```python
survey_length=1.0

##simulation file:
MC_var=np.loadtxt('Data/mc_var.dat').transpose()

Mabs_V_MC=MC_var[0] #M at the end of the plateau
s2_MC=MC_var[1]     #slope of the plateau
B_V_MC=MC_var[2]    #color at the end of the plateau
Pd_MC=MC_var[3]     #plateau duration
```

```python
#we will put these 1000 SNe at a random redshifts

N_SNe=len(Mabs_V_MC)

z_random=np.sort(np.random.random(100)*max(z_sim-0.04)+0.04)

m_SN=np.zeros((N_SNe,len(z_random)))
Pd_SN=np.zeros((N_SNe,len(z_random)))
delta_t_SN=np.zeros((N_SNe,len(z_random)))

for i,zs in enumerate(z_random):
    m_SN[:,i]=mu_LCDM(zs)+Mabs_V_MC+Kxr_func(zs)
    Pd_SN[:,i]=Pd_MC*(1+zs)
```

This is how the 'distribution' of apparent magnitudes AT THE END OF THE PLATEAU PHASE for the first 2000 (over 1.000,000) SNe II look at differnet redshifts. 

Those 'jumps' happen at redshifts where the best K-correction changes the observed filter.

```python
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')
for i in range(2000):
    ax1.plot(z_random,m_SN[i,:],alpha=0.9,linewidth=0.1,color='b')
ax1.set_xlabel(r'\textbf{redshift}')
ax1.set_ylabel(r'\textbf{m$_{end}$}')
ax1.set_title(r'\textbf{Distribution of m$_{end}$ for different z}')
plt.ylim(30,18)
plt.show()
```

Here we measure $\Delta_t(z)$, selecting different magnitude limit depending on the filter we use for observations.

```python
for j in range((N_SNe)):
    for i in range(len(z_random)):
        if z_random[i]<=0.3:           #V_swope would be converted to r_LSST                       
            if m_SN[j,i]<mag_lim['r']:
                delta_t_SN[j,i]=Pd_SN[j,i]
        elif (z_random[i]>0.3) and (z_random[i]<=0.57) :  #V_swope would be i_LSST
            if m_SN[j,i]<mag_lim['i']:
                delta_t_SN[j,i]=Pd_SN[j,i]
        elif (z_random[i]>0.57) and (z_random[i]<=0.78) : #V_swope would be z_LSST
            if m_SN[j,i]<mag_lim['z']:
                delta_t_SN[j,i]=Pd_SN[j,i]
        elif (z_random[i]>0.78) :      #V_swope would be y_LSST
            if m_SN[j,i]<mag_lim['y']:
                delta_t_SN[j,i]=Pd_SN[j,i]
```

NOTE: this may be improved by carrying the distribution of delta_t_SN at each redshift, instead of simply measuring the average at each redshift

```python
delta_t_SN_tot=np.zeros((len(z_random)))
for i in range(len(z_random)):
    delta_t_SN_tot[i]=sum(delta_t_SN[:,i])

delta_t_SN_tot=delta_t_SN_tot/N_SNe
```

```python
dV_dz_dW=np.zeros(len(z_random))

dn_dz_Horiuchi_CSP=np.zeros(len(z_random)) 
dn_dz_Cole_CSP=np.zeros(len(z_random))
dn_dz_Madau_CSP=np.zeros(len(z_random))

n_Horiuchi_CSP=np.zeros(len(z_random)) 
n_Cole_CSP=np.zeros(len(z_random))
n_Madau_CSP=np.zeros(len(z_random))

for i in range(len(z_random)):
    #dV(z) is the volume per unit redshift being surveyed
    dV_dz_dW[i]=(pow(r_com_LCDM(z_random[i]),2)*d_h*e_z_LCDM(z_random[i]))  
     #dn/dz is the SN per unit of redshift, comovil volume, solid angle and fied of view
    dn_dz_Horiuchi_CSP[i]=SNR_Horiuchi(z_random[i])/(1+z_random[i])*dV_dz_dW[i]*delta_omega
    dn_dz_Cole_CSP[i]    =SNR_Cole(z_random[i])    /(1+z_random[i])*dV_dz_dW[i]*delta_omega
    dn_dz_Madau_CSP[i]   =SNR_Madau(z_random[i])   /(1+z_random[i])*dV_dz_dW[i]*delta_omega

    #In yr (SNR in yr)
    n_Horiuchi_CSP[i]=dn_dz_Horiuchi_CSP[i]*(delta_t_SN_tot[i]/365.25)*survey_length
    n_Cole_CSP[i]=dn_dz_Cole_CSP[i]*(delta_t_SN_tot[i]/365.25)*survey_length
    n_Madau_CSP[i]=dn_dz_Madau_CSP[i]*(delta_t_SN_tot[i]/365.25)*survey_length
    
z_random,n_Horiuchi_CSP,n_Cole_CSP,n_Madau_CSP= zip(*sorted(zip(z_random,n_Horiuchi_CSP,n_Cole_CSP,n_Madau_CSP)))
fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')

ax1.plot(z_random,np.array(n_Cole_CSP),'k',z_random,np.array(n_Horiuchi_CSP),'g',z_random,np.array(n_Madau_CSP),'r')
ax1.set_title(r'\textbf{\# SNe II per unit of dz: sample simulated }')
ax1.set_xlabel(r'\textbf{redshift}')
ax1.set_ylabel(r'\textbf{\# SN }')
plt.legend((r'\textbf{Horiuchi}',r'\textbf{Cole}',r'\textbf{Madau}'))
plt.show()

print('Num. SNe Horiuchi: '+str(int(np.sum(n_Horiuchi_CSP))))
print('Num. SNe Cole:     '+str(int(np.sum(n_Cole_CSP))))    
print('Num. SNe Madau:     '+str(int(np.sum(n_Madau_CSP))))

z_random,n_Horiuchi_CSP,n_Cole_CSP,n_Madau_CSP = zip(*sorted(zip(z_random,n_Horiuchi_CSP,n_Cole_CSP,n_Madau_CSP)))
z_tot_Madau=[]
z_tot_Horiuchi=[]
z_tot_Cole=[]
for i,zs in enumerate(z_random):
    for j in range((int(round(n_Madau_CSP[i],0)))):
        z_tot_Madau.append(zs)
    for j in range((int(round(n_Cole_CSP[i],0)))):
        z_tot_Cole.append(zs)
    for j in range((int(round(n_Horiuchi_CSP[i],0)))):
        z_tot_Horiuchi.append(zs)
```

We have found the shape of the distribution, below one can create the final distribution by giving the total number of SNe II (with no selection effects) that LSST is supposed to found

```python
N_SNe_HD=10000#int(input("Number of SNe for the Hubble diagram: "))
z_SNe_HD_Madau=np.random.choice(z_tot_Madau, N_SNe_HD, replace=True)
z_SNe_HD_Horiuchi=np.random.choice(z_tot_Horiuchi, N_SNe_HD, replace=True)
z_SNe_HD_Cole=np.random.choice(z_tot_Cole, N_SNe_HD, replace=True)


fig,ax1 = plt.subplots(figsize=(9,6), facecolor='w', edgecolor='k')
ax1.hist(z_SNe_HD_Madau,color='red',alpha=0.9,density=True,edgecolor='red',ls='dashed',facecolor="None",lw=4,label=r'\textbf{Madau}')
ax1.hist(z_SNe_HD_Horiuchi,color='g',alpha=0.9,density=True,edgecolor='g',ls='dotted',facecolor="None",lw=4,label=r'\textbf{Horiuchi}')
ax1.hist(z_SNe_HD_Cole,color='k',alpha=0.9,density=True,edgecolor='k',facecolor="None",lw=4,label=r'\textbf{Cole}')
ax1.legend(loc=1,markerscale=0.5,prop={'size':12},ncol=2)
#plt.xlim(0,1.2)
ax1.set_xlabel(r'\textbf{redshift}')
ax1.set_ylabel(r'\textbf{SN}')
ax1.set_title(r'\textbf{SN II redshift distribution 1 yr: Single-visit}')
plt.show()
```
